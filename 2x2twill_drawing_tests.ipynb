```julia
function crossingAbstracted(start, vector0, vector1, (color0, color1, sz))
    if (sz == "Z")
#        setcolor(color0)
#        poly([start+vector0*(0, 0), start+vector0*(1, 1), start+vector0*(1, 1)+vector1, start+vector0*(0, 0)+vector1], 
#            close=true,  :fill)
        setcolor(color1)
        poly([start+2*vector1*(0, 0), start+2*vector1*(1, 1), start+2*vector1*(1, 1)+1*vector0, start+2*vector1*(0, 0)+1*vector0], 
            close=true,  :fill)
        setcolor("black")
        poly([start+2*vector1*(0, 0), start+2*vector1*(1, 1), start+2*vector1*(1, 1)+1*vector0, start+2*vector1*(0, 0)+1*vector0], 
            close=true,  :stroke)
    elseif (sz == "S")
#        setcolor(color1)
#        poly([start+vector1*(0, 0), start+vector1*(1, 1), start+vector1*(1, 1)+vector0, start+vector1*(0, 0)+vector0], 
#            close=true,  :fill)
        setcolor(color0)
        poly([start+2*vector0*(0, 0), start+2*vector0*(1, 1), start+2*vector0*(1, 1)+1*vector1, start+2*vector0*(0, 0)+1*vector1], 
            close=true,  :fill)
        setcolor("black")
        poly([start+2*vector0*(0, 0), start+2*vector0*(1, 1), start+2*vector0*(1, 1)+1*vector1, start+2*vector0*(0, 0)+1*vector1], 
            close=true,  :stroke)
    end
end
```

```julia
function crossing(start, vector0, vector1, (color0, color1, sz))
    outwidth=1.1
    boundwidth=.05
    inwidth=outwidth-boundwidth
    olap=0.2
    #olap=0.45 #overlap
    M = getmatrix()  #save current origin and scale
    translate(start[1], start[2])
    scale(1, 0.5)
    if (sz == "Z")
        #setcolor("black")
        #poly([vector0*(1-outwidth, 1-outwidth)-olap*vector1, 
        #        vector0*(outwidth, outwidth)-olap*vector1, 
        #        vector0*(outwidth, outwidth)+(1+olap)*vector1, 
        #        vector0*(1-outwidth, 1-outwidth)+(1+olap)*vector1], 
        #    close=true,  :fill)
        #setcolor(color0)
        #poly([vector0*(1-inwidth, 1-inwidth)-olap*vector1-vector1*(boundwidth, boundwidth),
        #        vector0*(inwidth, inwidth)-olap*vector1-vector1*(boundwidth, boundwidth), 
        #        vector0*(inwidth, inwidth)+(1+olap)*vector1*(1+boundwidth, 1+boundwidth), 
        #        vector0*(1-inwidth, 1-inwidth)+(1+olap)*vector1*(1+boundwidth, 1+boundwidth)], 
        #    close=true,  :fill)
        setcolor("black")
        poly([vector1*(1-outwidth, 1-outwidth)-olap*vector0, 
                vector1*(outwidth, outwidth)-olap*vector0, 
                vector1*(outwidth, outwidth)+(1+olap)*vector0, 
                vector1*(1-outwidth, 1-outwidth)+(1+olap)*vector0], 
            close=true,  :fill)
        setcolor(color1)
        poly([vector1*(1-inwidth, 1-inwidth)-olap*vector0-vector0*(boundwidth, boundwidth), 
                vector1*(inwidth, inwidth)-olap*vector0-vector0*(boundwidth, boundwidth), 
                vector1*(inwidth, inwidth)+(1+olap)*vector0*(1+boundwidth, 1+boundwidth), 
                vector1*(1-inwidth, 1-inwidth)+(1+olap)*vector0*(1+boundwidth, 1+boundwidth)], 
            close=true,  :fill)
    elseif (sz == "S")
        #setcolor("black")
        #poly([vector1*(1-outwidth, 1-outwidth)-olap*vector0, 
        #        vector1*(outwidth, outwidth)-olap*vector0, 
        #        vector1*(outwidth, outwidth)+(1+olap)*vector0, 
        #        vector1*(1-outwidth, 1-outwidth)+(1+olap)*vector0], 
        #    close=true,  :fill)
        #setcolor(color1)
        #poly([vector1*(1-inwidth, 1-inwidth)-olap*vector0-vector0*(boundwidth, boundwidth), 
        #        vector1*(inwidth, inwidth)-olap*vector0-vector0*(boundwidth, boundwidth), 
        #        vector1*(inwidth, inwidth)+(1+olap)*vector0*(1+boundwidth, 1+boundwidth), 
        #        vector1*(1-inwidth, 1-inwidth)+(1+olap)*vector0*(1+boundwidth, 1+boundwidth)], 
        #    close=true,  :fill)
        setcolor("black")
        poly([vector0*(1-outwidth, 1-outwidth)-olap*vector1, 
                vector0*(outwidth, outwidth)-olap*vector1, 
                vector0*(outwidth, outwidth)+(1+olap)*vector1, 
                vector0*(1-outwidth, 1-outwidth)+(1+olap)*vector1], 
             close=true,  :fill)
        setcolor(color0)
        poly([vector0*(1-inwidth, 1-inwidth)-olap*vector1-vector1*(boundwidth, boundwidth), 
                vector0*(inwidth, inwidth)-olap*vector1-vector1*(boundwidth, boundwidth), 
                vector0*(inwidth, inwidth)+(1+olap)*vector1*(1+boundwidth, 1+boundwidth), 
                vector0*(1-inwidth, 1-inwidth)+(1+olap)*vector1*(1+boundwidth, 1+boundwidth)], 
            close=true,  :fill)
      end
    setmatrix(M) #restore origin and scale
    #origin()
end
```

```julia
function reverseCrossing(start, vector0, vector1, (color1, color0, sz))
    if (sz == "S")
        setcolor(color0)
        poly([start+vector0*(.2, .2), start+vector0*(.8, .8), start+vector0*(.8, .8)+vector1, start+vector0*(.2, .2)+vector1], 
            close=true,  :fill)
        setcolor(color1)
        poly([start+vector1*(.2, .2), start+vector1*(.8, .8), start+vector1*(.8, .8)+vector0, start+vector1*(.2, .2)+vector0], 
            close=true,  :fill)
        setcolor("black")
        poly([start+vector1*(.2, .2), start+vector1*(.8, .8), start+vector1*(.8, .8)+vector0, start+vector1*(.2, .2)+vector0], 
            close=true,  :stroke)
    elseif (sz == "Z")
        setcolor(color1)
        poly([start+vector1*(.2, .2), start+vector1*(.8, .8), start+vector1*(.8, .8)+vector0, start+vector1*(.2, .2)+vector0], 
            close=true,  :fill)
        setcolor(color0)
        poly([start+vector0*(.2, .2), start+vector0*(.8, .8), start+vector0*(.8, .8)+vector1, start+vector0*(.2, .2)+vector1], 
            close=true,  :fill)
        setcolor("black")
        poly([start+vector0*(.2, .2), start+vector0*(.8, .8), start+vector0*(.8, .8)+vector1, start+vector0*(.2, .2)+vector1], 
            close=true,  :stroke)
    end
end
```

```julia
function makeLabel(n, odds, evens)
     string(join([odds[i]==1 ? string(2*n+1 - 2*i, ".") : "" for i=n:-1:1]), "/", 
        join([evens[i]==1 ? string(2*i, ".") : "" for i=1:n]))
end
```

```julia
function fillGrid(n, left, right, colors)
    grid = Array{String}(undef, n, n, 3)
    for i = 1:n
        for j = 1:n
            grid[i,j,1] = colors[left[j%n+1]+1]
            grid[i,j,2] = colors[right[i%n+1]+1]
            if (i+j) % 4 == 0
                grid[i,j,3] = "S"
            elseif (i+j) % 4 == 2
                grid[i,j,3] = "Z"
            else
                grid[i,j,3] = "N"
            end
        end
    end
    print(makeLabel(n, left, right)*"\n")
    return grid
end
```

```julia
function drawNumberedGrid(n, grid, start, vector0, vector1)
if grid != nothing
    for i = -n÷2:3*n÷2
        for j = max(-i, i-n+1):min(2*n-i, i+n)
            crossing(start+vector0*(i,i)+vector1*(j,j), vector0, vector1, Tuple(grid[(i+16*n)%2n+1,(j+16*n)%2n+1,:]))
            setcolor("black")
            #setmatrix([1, 0, 0, 1, 0, 0])
            #origin()    
            if (i+j+2) % 2 == 1
                text(string(2*(((3j+i+2n+4)÷2)%n+1)), start+vector0*(i+0.25,i+0.25)+vector1*(j+0.25,j+0.25), halign=:center, valign=:middle) 
            else
                text(string(2*((4n+2-((3i+j+3*n)÷2))%n+1)+1), start+vector0*(i+0.25,i+0.25)+vector1*(j+0.25,j+0.25), halign=:center, valign=:middle) 

            end
        end
    end
end
end
```

```julia
function drawAbstractedGrid(n, grid, start, vector0, vector1)
if grid != nothing
    for i = -n÷2:3*n÷2
        for j = max(-i, i-n+1):min(2*n-i, i+n)
            crossingAbstracted(start+vector0*(i,i)+vector1*(j,j), vector0, vector1, Tuple(grid[(i+3*n)%n+1,(j+3*n)%n+1,:]))
        end
    end
end
end
      
```

```julia
function drawNumberedAbGrid(n, grid, start, vector0, vector1)
if grid != nothing
    for i = -n÷2:3*n÷2
        for j = max(-i, i-n+1):min(2*n-i, i+n)
            crossingAbstracted(start+vector0*(i,i)+vector1*(j,j), vector0, vector1, Tuple(grid[(i+3*n)%n+1,(j+3*n)%n+1,:]))
            if (i+j+2) % 4 == 0 #odd
                setcolor("black")
                text(string(2*((-j+3*n-2)%n)+1), start+vector0*(i+0.5,i+0.5)+vector1*(j+0.5,j+0.5), halign=:center, valign=:middle) 
            elseif (i+j+2) % 4 == 2 #even
                  text(string(2*((i-3+3*n÷2)%n+1)), start+vector0*(i+0.5,i+0.5)+vector1*(j+0.5,j+0.5), halign=:center, valign=:middle) 
              # text(string(i)*","*string(j), start+vector0*(i+0.5,i+0.5)+vector1*(j+0.5,j+0.5), halign=:center, valign=:middle) 

            end
        end
    end
end
end
      
```

```julia
function drawLabeledAbGrid(n, grid, start, vector0, vector1)
    drawNumberedAbGrid(n, grid, start, vector0, vector1)
    for i = 1:16
#        text(string(i), O+((i-7.5)unit, -10unit), halign=:center, valign=:middle)
        label(string(i), :N, O+((i-7.5)unit, -9.5unit), leader=true, offset=1unit)
    end
end
```

```julia
function drawGrid(n, grid, start, vector0, vector1)

if grid != nothing
    for i = -n÷2:3*n÷2
        for j = max(-i, i-n+1):min(2*n-i, i+n)
            crossing(start+vector0*(i,i)+vector1*(j,j), vector0, vector1, Tuple(grid[(i+16*n)%2n+1,(j+16*n)%2n+1,:]))
        end
    end
end
end            
```

```julia
function evertGrid(n, grid, start, vector0, vector1)
if grid != nothing
    for i = 1:n
        for j = 1:n
            reverseCrossing(start+vector0*(i, i)+vector1*(j,j), vector0, vector1, Tuple(grid[j, i,:]))
        end
    end
end
end
```
